```python
# (   )
duplicated_rows = df.head(3)
zero_row = pd.DataFrame([[0] * len(df.columns)], columns=df.columns)
df = pd.concat([df, duplicated_rows, zero_row], ignore_index=True)
```

```python
def duplicates_nan(df, is_train=True):
    print(f' ,   : {df.isnull().any().sum()}')
    print(f'  : {df.isnull().values.sum()}')
    
    zero_rows = df[(df == 0).all(axis=1)]
    print(f'  : {zero_rows.shape[0]}')
    
    duplicates = df[df.duplicated()]
    print(f': {duplicates}')
    
    df = df[~(df == 0).all(axis=1)]
    
    if df.isnull().values.sum() > 0: #      0
        df = df.dropna(how='any') #   ,     .
    
    if df.duplicated().sum() > 0: #     0
        df = df.drop_duplicates(keep='first') #   ,       (keep='first'),      (inplace=True).
    
    return df
df = duplicates_nan(df)
```

```python
columns_scale = ['age', 'height(cm)', 'weight(kg)', 'waist(cm)', 'eyesight(left)',
       'eyesight(right)', 'systolic',
       'relaxation', 'fasting blood sugar', 'Cholesterol', 'triglyceride',
       'HDL', 'LDL', 'hemoglobin', 'serum creatinine', 'AST',
       'ALT', 'Gtp']
```

```python
def standardizing(dataframe, is_train=True):
    scaler = StandardScaler()
    if is_train:
        dataframe[columns_scale] = scaler.fit_transform(dataframe[columns_scale])
    else:
        dataframe[columns_scale] = scaler.transform(dataframe[columns_scale])
    return dataframe
```

```python
stand_df = standardizing(df)
```
